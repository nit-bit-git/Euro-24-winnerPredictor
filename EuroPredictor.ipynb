```python
import pandas as pd
```

```python
def transform_match_data(name):
    parts = name.split(" " , 1)
    type(parts)
    team_name = parts[1] if len(parts)>1 else parts[0]
    
    if team_name == 'N. Macedonia':
        team_name = 'North Macedonia'
    return f"{team_name} Men"
def correct_ga_gf(ga, gf):
    parts= ga.split("(", 1)
    parts2=gf.split("(", 1)
    parts[0] = parts[0].rstrip(" ") if len(parts) > 1 else parts[0]
    parts2[0] = parts2[0].rstrip(" ") if len(parts2) > 1 else parts2[0]
    ga = ".".join([parts[0],parts[1].strip().rstrip(")")]) if len(parts)>1 else parts[0]
    gf = ".".join([parts2[0], parts2[1].strip().rstrip(")")]) if len(parts)>1 else parts2[0]
    return ga, gf
matches["opponent"] = matches["opponent"].apply(transform_match_data)
matches["ga"] = matches["ga"].apply(lambda x: correct_ga_gf(x, x)[0])
matches["gf"] = matches["gf"].apply(lambda x: correct_ga_gf(x, x)[1])

matches.to_csv('matches2.csv')

print("Transformation complete. Check 'matches2.csv' for the result.")
    
    
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team", as_index=True).apply(lambda x: rolling_averages(x, cols, new_cols))
```
